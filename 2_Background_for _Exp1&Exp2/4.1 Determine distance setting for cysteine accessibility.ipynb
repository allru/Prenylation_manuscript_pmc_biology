```python
# Load data
datafolder = 'data'
PSSMSearch = 'data\PSSMSearch'
figures = 'data/figures'

# Background proteins with 1 cysteine
intern_Ccount_1 = pd.read_csv(os.path.join(datafolder, 'total_extracts_GG_F', 'background_C1.csv'), sep=',')
```

```python
len(intern_Ccount_1)
```

```python
# Define the list of distances to test
distances = [24, 30, 35, 40, 45, 50, 55, 60]

# Iterate over the list of distances to create annotated outputs

for dist in distances:
    group_name = f'{dist}tests_identified_C1'
    
    # Dynamically create a variable with the name C1_{dist}
    exec(f'C1_{dist} = annotate_structuremap(intern_Ccount_1, "up_output_background_multi_disulfid_17.04.24.csv", group=group_name, max_dist={dist}, max_angle=180)')
```

```python
# Merge all tests

data_frames = [C1_24, C1_30, C1_35, C1_40, C1_45, C1_50, C1_55, C1_60]

internC1_tests = reduce(lambda  left,right: pd.merge(left,right,
                                            how='inner'), data_frames)
```

```python
len(internC1_tests)
```

```python
# 74 proteins instead of the initial 75 because 1 doesn't have the same cysteine position in AlphaFold as determined previously

intern_Ccount_1[~intern_Ccount_1['ID'].isin(internC1_tests['ID'])]
```

```python
# Count which distance setting yields how many neighbours in total

# sort proteins by total cysteines (including those in disulfide bonds)
internC1_tests = internC1_tests.sort_values('Count_all').reset_index(drop=True)

# show only the total of neighbours per pPSE value (distance)
internC1_tests.loc['Total'] = internC1_tests.iloc[:, -8:].sum()
tests = pd.DataFrame(internC1_tests.iloc[-1, -8:]).transpose()
tests
```
